```python
# seperating the data for analysis
legit_transactions = credit_card_dataset[credit_card_dataset.Class == 0]
fraud_transactions = credit_card_dataset[credit_card_dataset.Class == 1]
```

Under-Sampling

Build a sample dataset containing similar distribution of normal transaction and fraudulent transaction

Number of Fraudulent Transactions = 492

```python
legit_sample = legit_transactions.sample(n=492)
```

Concatanating two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud_transactions], axis=0)
```

```python
new_dataset.head()
```

```python
print(new_dataset.shape)
```

```python
new_dataset.groupby("Class").mean()
```

```python
# Let's Split the data
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
print(Y)
```

```python
# Split the data to training and test data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =0.2, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# lets train the model
model = LogisticRegression()
```

```python
#Training the Logistic Reggression Model with training data
model.fit(X_train, Y_train)
```

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data: ',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on training data: ',test_data_accuracy)
```
